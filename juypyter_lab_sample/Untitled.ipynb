```python
from sklearn.linear_model import LinearRegression
X = df[['教育費']]
y = df['県民取得']
model = LinearRegression()
model.fit(X, y)  
```

```python
#教育費が20,000円のときの県民取得を予測
import numpy as np
model.predict(np.array([[50000]]))
```
